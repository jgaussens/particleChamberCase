```python
import re

import time
```

```python
class particle:
    def __init__(self, position, speed, direction):
        
        
        self.position = position
        
        if (direction != "R" and direction != "L"):
            raise ValueError("directions should be R or L only")

        
        if direction == "R":
            self.speed = speed
        if direction == "L":
            self.speed = -speed
            
        self.direction = direction
   

    def getParticleNextPosition(self):
        return self.position + self.speed
    

 
```

```python
class animation:
        
    def __init__(self, stringOfParticle, speed):
        
        self.size = len(stringOfParticle)
        self.speed = speed
        
        self.arrayOfParticle = list(stringOfParticle)
        
        for i in range(self.size):
            if stringOfParticle[i] != ".":
                self.arrayOfParticle[i] = particle(i, speed, stringOfParticle[i])
    
    
    
    #returns next position of the board depending on its previous state and speed
    def getNextPosition(self):
        board = ""
        for i in range(self.size):
            board += "."
        board = list(board)
        
            
        for particle in self.arrayOfParticle: #we parse each particle in the whole string and calculate the new position
            try:
                pos = particle.getParticleNextPosition()

                if(pos >= 0 and pos < self.size):
                    board[pos] = particle.direction
            except:
                continue


        return(board)
    
    def xPrints(leftString, rightString): #tool function to print desired output
        output = ""
        for i in range(len(leftString)):
            if (leftString[i] == "L" or rightString[i] == "R"):
                output += "X"
            else:
                output += "."
        print(output)
        
    
    def animate(speed, init):
        
        #assertions to fit requirements
        assert (speed >= 1 and speed <= 10), "Speed must be between 1 and 10"
        assert (len(init) >= 1 and len(init) <= 50), "init must contain between 1 and 50 characters inclusive"
        assert init and all(c in ".LR" for c in init), "init string is invalid. must be L, R or ."


        leftStr = str.replace(init, "R", ".")
        rightStr = str.replace(init, "L", ".")
        
        leftInit = list(leftStr)
        rightInit = list(rightStr)
        
        print(str.replace(str.replace(init, "L", "X"), "R", "X")) #desired output line by line
        
        
        while ( not all(v == "." for v in leftStr) or not all(v == "." for v in rightStr)):     #while there are still particles on board     
            
            left = animation(leftStr, speed) #create animation to given string
            leftInit = animation.getNextPosition(left) #get next position
            leftStr = "".join(leftInit) #transform into string to iterate again
            
            right = animation(rightStr, speed)
            rightInit = animation.getNextPosition(right)
            rightStr = "".join(rightInit)
            
            
            animation.xPrints(leftInit, rightInit)

            
            
```

```python
print("animating 2, ..R....")
animation.animate(2, "..R....")
print("")
print("")


print("animating 3, RR..LRL")
animation.animate(3, "RR..LRL")

print("")
print("")


print("animating 2, LRLR.LRLR")
animation.animate(2, "LRLR.LRLR")

print("")
print("")

print("animating 1, ...")

animation.animate(1, "...")


print("")
print("")


print("animating 10, RLRLRLRLRL")

animation.animate(10, "RLRLRLRLRL")

print("")
print("")

print("animating 1, LRRL.LR.LRR.R.LRRL.")
animation.animate(1, "LRRL.LR.LRR.R.LRRL.")

```
